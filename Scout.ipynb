```python
nan_linhas = argentina['Nação'].isnull()
nan_argentina = argentina.loc[nan_linhas]
nan_argentina
```

```python
argentina['Nação'] = argentina['Nação'].replace('af AFG', 'AFEGANISTAO')
argentina['Nação'] = argentina['Nação'].replace('ch SUI', 'SUICA')
argentina['Nação'] = argentina['Nação'].replace('bo BOL', 'BOLIVIA')
argentina['Nação'] = argentina['Nação'].replace('us USA', 'ESTADOS UNIDOS')
argentina['Nação'] = argentina['Nação'].replace('ar ARG', 'ARGENTINA')
argentina['Nação'] = argentina['Nação'].replace('si SVN', 'ESLOVENIA')
argentina['Nação'] = argentina['Nação'].replace('py PAR', 'PARAGUAI')
argentina['Nação'] = argentina['Nação'].replace('co COL', 'COLOMBIA')
argentina['Nação'] = argentina['Nação'].replace('pe PER', 'PERU')
argentina['Nação'] = argentina['Nação'].replace('uy URU', 'URUGUAI')
argentina['Nação'] = argentina['Nação'].replace('cl CHI', 'CHILE')
argentina['Nação'] = argentina['Nação'].replace('it ITA', 'ITALIA')
argentina['Nação'] = argentina['Nação'].replace('ec ECU', 'EQUADOR')
argentina['Nação'] = argentina['Nação'].replace('am ARM', 'ARMENIA')
argentina['Nação'] = argentina['Nação'].replace('ve VEN', 'VENEZUELA')
```

```python
mask = argentina.isnull().any(axis=1)
resultado = argentina[mask]
print(resultado)
```

```python
argentina = argentina.fillna(value=0)
```

```python
argentina['Nascimento'] = argentina['Nascimento'].astype(int)
argentina['Idade'] = argentina['Idade'].astype(int)
argentina.head()
```

```python
argentina.columns = ["Id", "Jogador", "Nação", "Posicao", "Equipe", "Idade", "Nascimento", "Jogos", "Iniciados", "Minutos",
              "Minutos/90", "Gols", "Assistencias", "Gols+Assistencias", "Gols_s/penaltis", "Gols_penaltis", "Penaltis_batidos", "Amarelos", "Vermelhos",
              "Gols/jogo", "Ass/jogo", "Gols+Ass/jogo", "Gols_s/penaltis/jogo", "Gols+Ass_s/penaltis/jogo"]
```

```python
argentina.head()
```

```python
argentina['Posicao'] = argentina['Posicao'].replace('AT', 'Atacante')
argentina['Posicao'] = argentina['Posicao'].replace('G', 'Goleiro')
argentina['Posicao'] = argentina['Posicao'].replace('ZG', 'Defesa')
argentina['Posicao'] = argentina['Posicao'].replace('LT', 'Meio-campo')
argentina['Posicao'] = argentina['Posicao'].replace('ATLT', 'Meia-Atacante')
argentina['Posicao'] = argentina['Posicao'].replace('LTAT', 'Meia-Atacante')
argentina['Posicao'] = argentina['Posicao'].replace('ZGLT', 'Defesa')
argentina['Posicao'] = argentina['Posicao'].replace('LTZG', 'Defesa')
```

```python
argentina['Posicao'].value_counts()
```

```python
argentina1.columns = ["Id", "Jogador", "Nação", "Posicao", "Equipe", "Idade", "Nascimento", "Jogos", "Gols", "Total_chutes",
              "Chutes_a_gol", "Chutes_acertos", "Chutes/jogos", "Chutes_acertos/jogos", "Gols/chutes", "Gols/chutes_acertos", "dist", "Gols_penaltis", "Penaltis_batidos"]
argentina1.head()              
```

```python
argentina2.columns = ["Id", "Jogador", "Nação", "Posicao", "Equipe", "Idade", "Nascimento", "Jogos", "Amarelos", "Vermelhos",
              "Segundo_amarelo", "Faltas_cometidas", "Faltas_sofridas", "Impedimentos", "Cruzamentos", "Cortes", "Roubada_bola", "penaltis", "penaltis1",
              "Gol_contra"]
argentina2.head()
```

```python
jogadores_arg = pd.concat([argentina, argentina1, argentina2], axis=1, join='outer')

# remover as colunas duplicadas
jogadores_arg = jogadores_arg.loc[:,~jogadores_arg.columns.duplicated()]

# exibir o dataframe resultante
jogadores_arg.head()


```

```python
jogadores_arg = jogadores_arg.drop(['penaltis', 'penaltis1', 'dist'], axis=1)
jogadores_arg.info()
```

```python
jogadores_arg.head()
```

```python
jogadores_arg_sub23 = jogadores_arg.query("Idade<=23")
jogadores_arg_sub23.sort_values('Idade').head()
```

```python
jogadores_sem_idade = jogadores_arg_sub23.query("Idade==0")
jogadores_sem_idade
```

```python
jogadores_arg_sub23 = jogadores_arg_sub23.drop(jogadores_arg_sub23[jogadores_arg_sub23['Idade'] == 0].index)
jogadores_arg_sub23.head()
```

```python
jogadores_arg_sub23.info()
```

```python
jogadores_sub23 = jogadores_arg_sub23.query("Jogos>=16")
jogadores_sub23.sort_values('Jogos').head()
```

```python
jogadores_sub23.loc[:,'Ass/jogo'] = jogadores_sub23['Ass/jogo'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Gols+Ass/jogo'] = jogadores_sub23['Gols+Ass/jogo'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Gols_s/penaltis/jogo'] = jogadores_sub23['Gols_s/penaltis/jogo'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Gols+Ass_s/penaltis/jogo'] = jogadores_sub23['Gols+Ass_s/penaltis/jogo'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Chutes/jogos'] = jogadores_sub23['Chutes/jogos'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Chutes_acertos/jogos'] = jogadores_sub23['Chutes_acertos/jogos'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Gols/chutes'] = jogadores_sub23['Gols/chutes'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Gols/chutes_acertos'] = jogadores_sub23['Gols/chutes_acertos'].apply(lambda x: round(x / 100, 2))
jogadores_sub23.loc[:,'Chutes_acertos'] = jogadores_sub23['Chutes_acertos'].apply(lambda x: round(x / 10, 1))
```

```python
jogadores_sub23.loc[:,'Gols/jogo'] = jogadores_sub23['Gols/jogo'].apply(lambda x: round(x / 100, 2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Criando o gráfico de dispersão
ax = jogadores_sub23.plot.scatter(x='Total_chutes', y='Chutes_a_gol')

# Adicionando o nome do jogador para cada ponto no gráfico
for i, row in jogadores_sub23.iterrows():
    ax.annotate(row['Jogador'], (row['Total_chutes'], row['Chutes_a_gol']))

# Exibindo o gráfico
plt.show()
```

```python
# classificar o DataFrame com base nas pontuações de gols e assistências
jogadores_sub23 = jogadores_sub23.sort_values(by=['Gols', 'Assistencias'], ascending=False)

# filtrar o DataFrame para incluir apenas os primeiros 10 jogadores
jogadores_sub23 = jogadores_sub23.head(10)

# plotar o gráfico de barras empilhadas
plt.figure(figsize=(16,3))
sns.barplot(x='Jogador', y='Assistencias', data=jogadores_sub23, palette='dark')
sns.barplot(x='Jogador', y='Gols', data=jogadores_sub23, palette='bright')


# personalizar o gráfico
plt.title('Gols e assistências por jogador')
plt.xlabel('Jogador')
plt.ylabel('Número de gols ou assistências')
plt.show()
```
