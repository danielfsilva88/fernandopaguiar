```python
import pandas as pd
from statsbombpy import sb
from mplsoccer import VerticalPitch, Pitch
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
from highlight_text import ax_text, fig_text
```

```python
sb.competitions()
```

```python
wc = sb.matches(competition_id=72, season_id=30)
wc
```

```python
events = sb.events(match_id="69321")
events.info()
```

```python
events['type'].unique()
```

```python
passes = events[events['type'] =='Pass']
passes
```

```python
passes.columns
```

```python
passes['location']
```

```python
passes[['x', 'y']] = passes['location'].apply(pd.Series)
passes[['pass_end_x', 'pass_end_y']] = passes['pass_end_location'].apply(pd.Series)
passes
```

```python
passes_eua = passes[(passes['team']=="United States Women's")]
passes_eua
```

```python
passes_eua['player'].unique()
```

```python
passes_total = passes_eua.groupby(['player'])['player'].count().to_frame()
passes_total
```

```python
passes_eua['pass_outcome'].unique()
```

```python
completos = passes_eua[passes_eua['pass_outcome'].isnull()]
incompletos = passes_eua[passes_eua['pass_outcome'].notnull()]
completos
```

```python
passes_total['completos'] = completos.groupby(['player'])['player'].count().to_frame()
passes_total['incompletos'] = incompletos.groupby(['player'])['player'].count().to_frame()
passes_total = passes_total.fillna(0)
passes_total
```

```python
passes_total = passes_total.rename(columns={'player': 'passes total'})
passes_total['Porcentagem %'] = passes_total['completos'] / passes_total['passes total'] * 100
passes_total = passes_total[passes_total['passes total'] >= 10]
passes_total
```

```python
passes_total = passes_total.reset_index()
passes_total
```

```python
top_10 = passes_total.sort_values('Porcentagem %', ascending = True).tail(10).reset_index()
top_10 = top_10.drop('index', axis=1)
top_10
```

```python
fig = plt.figure(figsize=(8,6))
ax = plt.axes()
ax.barh(top_10['player'],
        top_10['Porcentagem %'])
plt.title("Melhor porcentagem de passes na Final")
plt.xlabel('Porcentagem de passes')
plt.ylabel('Jogadora')
plt.show()
```

```python
ax = passes_total[['player', 'completos', 'incompletos']].sort_values('completos', ascending=True).head(10).plot.barh(x='player', stacked=True)
plt.title('Passes vs Holanda')
plt.xlabel('% Passes')
plt.ylabel('Jogadora')
plt.show()
```

```python
white='white'
sbred='#e21017'
lightgrey='#d9d9d9'
darkgrey='#9A9A9A'
cmaplist = [white, darkgrey, sbred]
cmap = LinearSegmentedColormap.from_list("",cmaplist)
```

```python
def passMap(player):

  player=player

  comp=completos[(completos['player']==player)]
  incomp=incompletos[(incompletos['player']==player)]

  pitch = Pitch(pitch_type="statsbomb", pitch_color='white', line_color='black', line_zorder=2)

  fig, ax = pitch.draw(figsize=(13,8), constrained_layout=True, tight_layout=False)
  fig.set_facecolor('white')

  bin_statistic = pitch.bin_statistic(comp.pass_end_x, comp.pass_end_y, statistic='count', bins=(12,8), normalize=True)
  pitch.heatmap(bin_statistic, ax=ax, alpha=0.5, cmap=cmap)

  pitch.arrows(comp.x, comp.y, comp.pass_end_x, comp.pass_end_y, width=3, headwidth=8, headlength=5, color=sbred, ax=ax, zorder=2, label='Passes completos')
  pitch.arrows(incomp.x, incomp.y, incomp.pass_end_x, incomp.pass_end_y, width=3, headwidth=8, headlength=5, color=darkgrey, ax=ax, zorder=2, label='Passes incompletos')

  ax.legend(facecolor='white', handlelength=5, edgecolor='None',fontsize=20, loc='best')

  ax_title = ax.set_title('Passes de ' f'{player} vs Holanda', fontsize=10, color='black')

  plt.show()
```

```python
passMap('Abby Dahlkemper')
```

```python
for player in passes_total['player'].head().unique():
  passMap(player)
```
